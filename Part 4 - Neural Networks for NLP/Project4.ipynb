```python
import requests
from colorama import Fore, Style

API_URL = "https://api-inference.huggingface.co/models/AdapterHub/bert-base-uncased-pf-ud_pos"
headers = {"Authorization": "Bearer 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"}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

output = query({
    "inputs": "Overall , the code appears to be annotating and displaying text with its POS tags in a colorful and visually appealing",
})

# Initialize variables to keep track of the current word and POS tag
current_word = None
current_word_start = None
current_word_end = None
current_pos = None

# Process the output and combine subword tokens with complete words and their POS tags
complete_words_with_pos = []

for token in output:
    word = token['word']
    start = token['start']
    end = token['end']
    pos = token['entity_group']

    if current_word is None:
        current_word = word
        current_word_start = start
        current_word_end = end
        current_pos = pos
    else:
        if start == current_word_end:
            current_word += word
            current_word_end = end
        else:
            complete_words_with_pos.append({"word": current_word, "pos": current_pos})
            current_word = word
            current_word_start = start
            current_word_end = end
            current_pos = pos

# Append the last word
if current_word:
    complete_words_with_pos.append({"word": current_word, "pos": current_pos})

# Initialize color mapping for POS tags
pos_color_map = {
    "NOUN": Fore.GREEN,
    "PUNCT": Fore.WHITE,
    "ADP": Fore.CYAN,
    "NUM": Fore.MAGENTA,
    "SYM": Fore.YELLOW,
    "SCONJ": Fore.RED,
    "ADJ": Fore.BLUE,
    "PART": Fore.MAGENTA,
    "DET": Fore.CYAN,
    "CCONJ": Fore.RED,
    "PROPN": Fore.BLUE,
    "PRON": Fore.YELLOW,
    "X": Fore.WHITE,
    "_": Fore.MAGENTA,
    "ADV": Fore.GREEN,
    "INTJ": Fore.YELLOW,
    "VERB": Fore.CYAN,
    "AUX": Fore.RED,
}

colored_text = ""

# Assuming complete_words_with_pos is a list of word-POS tag pairs
for item in complete_words_with_pos:
    word = item['word']
    pos = item['pos']
    pos_color = pos_color_map.get(pos, Fore.RESET)  # Use Fore.RESET if the POS tag is not in the mapping
    colored_text += f"{pos_color}{word} ({pos}){Fore.RESET} "

# Print the formatted output
print(colored_text)
```
